```python
import matplotlib.pyplot as plt
import numpy as np

# Data: Paintings and Corresponding Years
paintings = ["The Two Fridas", "My Dress Hangs Here", "Tree of Hope",
             "Self Portrait With Monkeys", "The Broken Column",
             "The Wounded Deer", "Me and My Doll"]
dates = [1939, 1933, 1946, 1940, 1944, 1946, 1937]

# Sort paintings by year (optional)
paintings, dates = zip(*sorted(zip(paintings, dates)))

# Create Figure and Bar Chart
plt.figure(figsize=(10,5))
bars = plt.barh(paintings, dates, color='purple')

# Adjust X-axis for Better Scaling
plt.xlim(min(dates) - 5, max(dates) + 5)  # Adding padding for visibility

# Add Labels and Title
plt.xlabel("Year of Creation", fontsize=12)
plt.ylabel("Painting Name", fontsize=12)
plt.title("Frida Kahlo's Paintings by Year", fontsize=14)
plt.gca().invert_yaxis()  # Invert Y-axis for better readability

# Annotate Bars with Year Values
for bar, date in zip(bars, dates):
    plt.text(date - 1, bar.get_y() + bar.get_height()/2, str(date),
             va='center', ha='right', fontsize=10, color="white", weight="bold")

# Show the Updated Chart
plt.show()

```
